### Leia o csv e converta para dicionário

```python
consulta_PE_csv = pd.read_csv("consulta_cand_2020_PE_desafio.csv", encoding="latin_1", sep=",")

colunas = consulta_PE_csv.columns.values.tolist()
consulta_PE_csv.to_dict()
```

### (Mini) Análise de desempenho dos métodos

```python
# Utilizando o método shape()

print(consulta_PE_csv[consulta_PE_csv.NR_IDADE_DATA_POSSE < 23].shape[0])
%timeit consulta_PE_csv[consulta_PE_csv.NR_IDADE_DATA_POSSE < 23].shape[0]
```

```python
# Utilizando o método sum_mask_numpy
print((consulta_PE_csv.NR_IDADE_DATA_POSSE.values < 23).sum())
%timeit (consulta_PE_csv.NR_IDADE_DATA_POSSE.values < 23).sum()
```

###  Quantas pessoas candidatas no estado de Pernambuco seriam mais novas do que você 

```python
pessoas_mais_novas_do_que_eu = (consulta_PE_csv.NR_IDADE_DATA_POSSE.values < 23).sum()
pessoas_mais_novas_do_que_eu
```

### Quantos porcentos isso representa em relação ao total de pessoas candidatas.

```python
f"{pessoas_mais_novas_do_que_eu / len(consulta_PE_csv):.1%}"
```
